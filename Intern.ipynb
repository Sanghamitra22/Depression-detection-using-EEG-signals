```python
from scipy.integrate import simps
from scipy import signal
for files in glob.glob("/content/drive/MyDrive/Data2/Segmented/**"):
  low, high = 8, 13
  win = (2/low)*256
  

  final_list = []
  for j in range(2):
    data = pd.read_csv(files+'/'+'Segmented_'+ str(j) +'.csv')
    
    if len(data) !=0:
      result = []
      
      for i in range(20):
        data_1 = data.iloc[:, i]
        freqs, psd = signal.welch(data_1, 256, nperseg=win)
        # Find intersecting values in frequency vector
        alpha = np.logical_and(freqs >= low, freqs <= high)
        # Frequency resolution
        freq_res = freqs[1] - freqs[0]  

        # Compute the absolute power by approximating the area under the curve
        alpha_power = simps(psd[alpha], dx=freq_res)
        #print('Absolute alpha power: %.7f uV^2' % alpha_power)

        total_power = simps(psd, dx=freq_res)
        alpha_rel_power = alpha_power / total_power
        #print('Relative alpha power: %.3f' % alpha_rel_power)

        result.append(alpha_rel_power)
      final_list.append(result)
        
      data_frame = pd.DataFrame(final_list)
      if(len(data_frame.columns) == 20):
        data_frame.to_csv('/content/drive/MyDrive/Data2/Rel_S/'+'Rel_pwr_'+files[39:52]+'.csv', index=False, header=['Fp1-LE','F3-LE','C3-LE','P3-LE','O1-LE','F7-LE','T3-LE','T5-LE','Fz-LE','Fp2-LE','F4-LE','C4-LE','P4-LE','O2-LE','F8-LE','T4-LE','T6-LE','Cz-LE','Pz-LE','A2-A1'])
      elif(len(data_frame.columns) == 22):
        data_frame.to_csv('/content/drive/MyDrive/Data2/Rel_S/'+'Rel_pwr_'+files[39:52]+'.csv', index=False, header=['Fp1-LE','F3-LE','C3-LE','P3-LE','O1-LE','F7-LE','T3-LE','T5-LE','Fz-LE','Fp2-LE','F4-LE','C4-LE','P4-LE','O2-LE','F8-LE','T4-LE','T6-LE','Cz-LE','Pz-LE','A2-A1','23A-23R','24A-24R'])
    else:
      pass


```

```python
from tqdm import tqdm
import cv2
import os
dataset ='/content/drive/MyDrive/Data2/Train/'
images = []
labels = []
output=[]
IMAGE_SIZE = (64, 64)
print("Loading {}".format(dataset))

# Iterate through each folder corresponding to a category
for folder in os.listdir(dataset):
    label = class_names_label[folder]
    #print(folder)
    # Iterate through each image in our folder
    for file in tqdm(os.listdir(os.path.join(dataset, folder))):
        
        # Get the path name of the image
        img_path = os.path.join(os.path.join(dataset, folder), file)

        # Open and resize the img
        image = cv2.imread(img_path)
        #image = image[6:208,92:420]
        image = cv2.cvtColor(image, cv2.COLOR_BGR2RGB)
        image = cv2.resize(image, IMAGE_SIZE) 
        
        # Append the image and its corresponding label to the output
        images.append(image)
        labels.append(label)
        #cv2_imshow(image)
        #print(label)
images = np.array(images, dtype = 'float32')
labels = np.array(labels, dtype = 'int32')
#print("images=",images)
print("labels=",labels[194])

output.append((images, labels))
#print("output=",output)
from google.colab.patches import cv2_imshow
#for i in range(196):
#  cv2_imshow(images[i])
cv2_imshow(images[194])
outpath = "trial_MDD.png"
outpath2 = "trial_H.png"
#save the image
cv2.imwrite(outpath, images[194])
cv2.imwrite(outpath2, images[94])
cv2_imshow(images[94])
print(images[0].shape)
```

```python
import numpy as np
from sklearn.model_selection import train_test_split

X_train, X_test, y_train, y_test = train_test_split(images, labels, test_size=0.15, random_state=10)
```

```python

import keras
import tensorflow as tf
from keras.models import Sequential
from keras.layers.convolutional import Conv2D
from keras.layers import Dense
from keras.layers import Conv2D
from keras.layers import MaxPooling2D
from keras.layers import BatchNormalization
from keras.layers import Flatten
from keras.layers import Dropout
#def cnn():
model = Sequential()
model.add(Conv2D(32, (3, 3), input_shape=(64, 64, 3), activation='relu'))
model.add(MaxPooling2D(pool_size=(2, 2)))
#model.add(BatchNormalization())
model.add(Conv2D(64, (3, 3), input_shape=(32, 32, 3), activation='relu'))
model.add(MaxPooling2D(pool_size=(2, 2)))
model.add(Conv2D(128, (3, 3), input_shape=(32, 32, 3), activation='relu'))
model.add(MaxPooling2D(pool_size=(2, 2)))
model.add(Flatten())
model.add(Dense(256, activation='relu'))
model.add(Dropout(0.25))
model.add(Dense(1, activation='sigmoid'))
#model.summary()
opt = tf.keras.optimizers.Adam(learning_rate = 0.0001)
model.compile(
  optimizer= opt, 
  loss='binary_crossentropy', 
  metrics=['accuracy']
)
  #return model
history = model.fit(
  X_train, 
  y_train,
  validation_data=(X_test, y_test),
  epochs=100, 
  validation_split=0.15, 
  batch_size=1, 
  verbose=2
  
)
"""from sklearn.model_selection import StratifiedKFold
cv = StratifiedKFold(n_splits=5, random_state=0, shuffle=True)"""
# Evaluate accuracy on test and train sets
score_train = model.evaluate(X_train, y_train, verbose=1)
print('Train accuracy:', round(score_train[1],4))
score_test = model.evaluate(X_test, y_test, verbose=1)
print('Test accuracy:', round(score_test[1],4))
#print("Accuracy after fitting: {:.2f}%".format(history['accuracy'][-1]*100))
```

```python
for layer_name, feature_map in zip(layer_names, feature_maps):  
  if len(feature_map.shape) == 4:
      k = feature_map.shape[-1]  
      size=feature_map.shape[1]
      image_belt = np.zeros((size, size*k))
      for i in range(k):
        feature_image = feature_map[0, :, :, i]
        feature_image-= feature_image.mean()
        feature_image/= feature_image.std ()
        feature_image*=  64
        feature_image+= 128
        feature_image= np.clip(feature_image, 0, 255).astype('uint8')
        image_belt[:, i * size : (i + 1) * size] = feature_image
      scale = 20. / k
      plt.figure( figsize=(scale * k, scale) )
      plt.title ( layer_name, fontsize = 30 )
      plt.grid  ( False )
      plt.imshow( image_belt, aspect='auto',cmap='jet')
```
